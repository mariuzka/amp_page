## Aufgaben 1 - 15

```python
############################################################
# Module importieren (Aufgabe 1)
############################################################

# Das Modul *random* importieren
import random

############################################################
# Parameter festlegen (Aufgabe 2)
############################################################

# *TIME*, *STREET_LEN*, *DENSITY*, *MAX_SPEED* und *DAWDLING_PROB* festlegen
TIME = 200
STREET_LEN = 100
DENSITY = 4
MAX_SPEED = 5
DAWDLING_PROB = 0.1


############################################################
# Straße zusammenbauen und befüllen (Aufgabe 3)
############################################################

# leere Straßenliste erstellen
street = []

# für die Länge der Straße (*STREET_LEN*)
for i in range(STREET_LEN):
    
    # Wenn *i* modulo *DENSITY* gleich 0
    if i % DENSITY == 0:
        # *MAX_SPEED* an *street* anfügen (als Repräsentation eins Autos)
        street.append(MAX_SPEED)
    
    # Ansonsten:
    else:
        # *None* an *street* anfügen
        street.append(None)
    
    
############################################################
# Simulationsloop (Aufgabe 4)
############################################################

# für jeden Zeitschritt
for tick in range(TIME):
    
    
    ############################################################
    # graphische Darstellung (Aufgabe 5)
    ############################################################
    
    # leeren String namens *street_as_string* erstellen
    street_as_string = ""
    
    # für jedes Element der Liste *street*
    for element in street:
        
        # wenn das Element ein Integer d.h. ein Auto ist
        if type(element) == int:
            # Element in str umwandeln und an *street_as_string* anhängen
            street_as_string += str(element)
        
        # Ansonsten
        else:
            # einen Punkt anhängen
            street_as_string += "."
    
    # *street_as_string* printen
    print(street_as_string)
    
    
    ############################################################
    # neue Straße erstellen (Aufgabe 6)
    ############################################################
    
    # Straße für nächsten Zeitschritt erstellen
    new_street = []
    # für die Länge der Straße 
    for i in range(STREET_LEN):
        # *None* an *new_street* anhängen
        new_street.append(None)
    
    # Kurze Alternative 1:
    new_street = [None] * STREET_LEN
    
    # Kurze Alternative 2:
    new_street = [None for i in range(STREET_LEN)]
    
    
    ############################################################
    # Straßenliste durchgehen (Aufgabe 7)
    ############################################################
    
    # für jede Position der Straße
    for position in range(STREET_LEN):
        
        ############################################################
        # Aufgabe 8
        ############################################################
        
        # wenn das Element der Position ein Auto ist
        if type(street[position]) == int:
            
            ############################################################
            # Aufgabe 9
            ############################################################
            
            # Zur besseren Lesbarkeit aktuelle Geschwindigkeit unter *speed* abspeichern
            speed = street[position]
            
            
            ############################################################
            # 1. Handlungsregel (Aufgabe 10)
            ############################################################
            
            # Wenn *speed* kleiner *MAX_SPEED* d.h. wenn Agent nicht maximal schnell fährt
            if speed < MAX_SPEED:
                # Geschwindigkeit um 1 erhöhen
                speed += 1
            
            
            ############################################################
            # 2. Handlungsregel (Aufgabe 11)
            ############################################################
            
            # für jede beobachtsbare Position
            for i in range(speed):
                # aktuelle Beobachtungsdistanz berechnen
                visual_range = i + 1
                
                # Position aktuell beobachteter Position berechnen
                observed_position = (position + visual_range) % STREET_LEN
                
                # Sobald Element auf beobachteter Position ein Auto ist (d.h. ein int)
                if type(street[observed_position]) == int:
                    
                    # Geschwindigkeit so einstellen, dass man genau eine Zelle vor das nächste Auto fährt
                    speed = visual_range - 1
                    
                    # For-Loop abbrechen, da ein Hindernis gefunden wurde
                    break
                
            
            ############################################################
            # 3. Handlungsregel (Aufgabe 12)
            ############################################################
            
            # wenn *speed* größer 0 und Zufallswert zwischen 0 und 1 kleinergleich *DAWDLING_PROB*
            if speed > 0 and random.random() < DAWDLING_PROB:
                # *speed* um 1 verringern
                speed -= 1
            
            
            ############################################################
            # 4. Handlungsregel (Aufgabe 13)
            ############################################################
            
            # Neue Position berechnen
            new_position = (position + speed) % STREET_LEN
            
            # Agent d.h. seine Repräsentation als Geschwindigikeit auf neue Straße setzen
            new_street[new_position] = speed
    
    
    ############################################################
    # Straße aktualisieren (Aufgabe 14)
    ############################################################
    
    # *new_street* wird die neue *street*
    street = new_street
```
